```python
# remove constant columns
colsToRemove = []
for col in data.columns:
    if data[col].std() == 0:
        colsToRemove.append(col)

data.drop(colsToRemove, axis=1, inplace=True)

# remove duplicate columns
colsToRemove = []
columns = data.columns
for i in range(len(columns)-1):
    v = data[columns[i]].values
    for j in range(i+1,len(columns)):
        if np.array_equal(v,data[columns[j]].values):
            colsToRemove.append(columns[j])

data.drop(colsToRemove, axis=1, inplace=True)

```

SAMPLING FOR CLASS IMBALANCE

```python
#SMOTE to Upsample unsatisied customers
X = data.loc[:, data.columns != 'TARGET']
y = data.loc[:, data.columns == 'TARGET']
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y.values.ravel(), test_size=0.3, random_state=0)
columns = X_train.columns
os_data_X,os_data_y=os.fit_sample(X_train, y_train)
os_data_X = pd.DataFrame(data=os_data_X,columns=columns )
os_data_y= pd.DataFrame(data=os_data_y, columns=['y'])
# we can Check the numbers of our data
print("length of oversampled data is ",len(os_data_X))
print("Number of satisfied customers in oversampled data",len(os_data_y[os_data_y['y']==0]))
print("Number of unsatisfied customers",len(os_data_y[os_data_y['y']==1]))
print("Proportion of satisfied customers in oversampled data is ",len(os_data_y[os_data_y['y']==0])/len(os_data_X))
print("Proportion of unsatisfied customers in oversampled data is ",len(os_data_y[os_data_y['y']==1])/len(os_data_X))
```

XGBOOST Classifier

```python
# instantiate and fit XGBClassifier
clf = XGBClassifier()
clf.fit(os_data_X, os_data_y)

# predict on training and test sets
training_preds = clf.predict(os_data_X)
test_preds = clf.predict(X_test)

# accuracy of training and test sets
training_accuracy = accuracy_score(os_data_y, training_preds)
test_accuracy = accuracy_score(y_test, test_preds)

print('Training Accuracy: {:.4}%'.format(training_accuracy * 100))
print('Validation Accuracy: {:.4}%'.format(test_accuracy * 100))
```

```python
submission = pd.DataFrame({"ID":X_test.index, "TARGET":test_preds})
submission.TARGET.value_counts(0)
```

```python
train=pd.concat([X_train_new,y_train],axis=1)
# find and remove correlated features
def correlation(dataset, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr

corr_features = correlation(train, 0.9)
print('correlated features: ', len(set(corr_features)) )
```

```python
# removed correlated  features
X_train_new.drop(labels=corr_features, axis=1, inplace=True)
X_test_new.drop(labels=corr_features, axis=1, inplace=True)

```
